### Questão 2:

Considerando a expressão,

$$\cfrac{e^{1/x}}{1+e^{1/x}} ~~~~(Eq. 1)$$

Dividindo tanto a expressão do denominador quanto a do numerador por $e^{1/x}$ temos:

$$\cfrac{e^{1/x}}{1+e^{1/x}} \div \cfrac{e^{1/x}}{e^{1/x}} = \cfrac {\cfrac{e^{1/x}}{e^{1/x}}}{\cfrac{1+e^{1/x}}{e^{1/x}}} = \cfrac{1}{\cfrac{1}{e^{1/x}}+1} = \boxed {~~~~\cfrac{1}{e^{-1/x}+1}~~~~} ~~~~(Eq. 2)$$



Agora usando o python para calcular a expressão 1 para valores pequenos de x temos:

```python
import math                                                      # biblioteca de matemática para usar a função exp()

print( "\n")
print ("Cálculo via Eq. 1\n")

for i in range (300, 0, -1):                                    # range (inicio, fim, passo)
    
    x = 0.0001*i
    print ("x =", x, )                                             # A vígula no final permite unir o próximo print à esta linha
    print (": ", math.exp(1.0/x)/(1 + math.exp(1.0/x)) )
```

Há um erro de overflow para a função exponencial quando x = 0.0014. 
Agora usando a expressão 2 temos:

```python
import math

print ("\n")
print ("Cálculo via eq.2 \n")

for i in range (300 , 0, -1):   
    
    x = 0.0001*i
    print ("x =", x, )                                       
    print (": ", 1/(math.exp(-1.0/x) + 1) )
```


Abaixo executamos alguns testes para expoentes positivos e negativos, e logo que o expoente for maior do que +709 a função retorna overflow.



```python
for i in range (700,800,1):
    print( i,)
    print( "    ",math.exp(i))
```

```python
for i in range (700,1000,1):
    print (i,)
    print (math.exp(-i))
```


Observando os resultados, vemos que, embora as duas equações sejam matematicamente idênticas, para $x$ menores que 0.0015, 
calcular a equação $\cfrac{e^{1/x}}{1+e^{1/x}}$ com a função math.exp() retorna um erro de overflow, pois nos termos $e^{1/x}$ o expoente vira um número positivo maior do que ela consegue calcular. Já na equação $1/(e^{-1/x}+1)$ a função math.exp() consegue receber números negativos de módulo maior.
